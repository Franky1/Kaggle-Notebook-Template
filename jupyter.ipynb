### Show Pairplot of Features

```python
# sns_pairplot = sns.pairplot(data=df, hue='label', palette='Set2')
sns_pairplot = sns.pairplot(data=df.sample(frac=0.1), kind="scatter", hue='label')
sns_pairplot.savefig("pairplot.png")
```

### Show Heatmap of Features

```python
plt.figure(figsize=(15, 15))
# sns_heatmap = sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt='.1f', cmap='RdYlGn', square=True)
sns_heatmap = sns.heatmap(df.corr(), annot=True, fmt='.1f', cmap='coolwarm', square=True)
sns_heatmap.get_figure().savefig("heatmap.png")
```
